```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('top_hiphop_artists_tracks.csv')
```

```python
df.head()
```

```python
df['Artist'].unique()
```

```python
df.isnull().sum()
```

```python
df_artist=df.groupby('Artist')['Popularity'].mean()
```

```python
most_popularity_artist = df_artist[df_artist > 75 ].sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.barplot(x=most_popularity_artist.index,y=most_popularity_artist)
plt.xticks(rotation=90);
plt.save('most_popularity_artist.png')
```
